```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_from_path(csv_path, xlim=None, ylim=None):
    df = pd.read_csv(csv_path)
    if xlim:
        plt.xlim(xlim)
    if ylim:
        plt.ylim(ylim)
    plt.plot(df['step'], df['mloss'], label='mloss')
    plt.plot(df['step'], df['gloss'], label='gloss')
    plt.xlabel('step')
    plt.ylabel('loss')
    plt.legend()
    plt.show()
```

```python
plot_from_path("baselines/tabddpm/ckpt/adult/step_loss_history.csv", ylim=(0,5))
```
